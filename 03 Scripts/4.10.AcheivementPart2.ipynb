```python
# Import libraries
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
# Importing dataframes
path = r'C:\Users\William Inglish\Instacart Basket Analysis'

df_final = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'final_df.pkl'))
```

```python
loyal_pieplot = df_final['loyalty_flag'].value_counts().plot.pie(title='Profile For Loyalty Flags',label='Loyalty Flags', shadow = True, autopct=('%1.1f%%'))
```

```python
# Creating a crosstab between loyalty flag and region
loyal_prior=pd.crosstab(df_final['days_since_prior_order'],df_final['loyalty_flag'],dropna=False)
```

```python
# Creating a bar chart between loyalty flag and region
loyal_prior_bar=loyal_prior.plot.barh(stacked=True,color=sns.color_palette('Blues',3))
plt.title('Profile Of Loyalty Flag', fontweight='bold')
plt.show()
```
