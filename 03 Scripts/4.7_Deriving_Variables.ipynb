## If Statement Using Define

```python
#Creating a definition for price labeling
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
#Imputing new column with def price_label

df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna= False)
```

```python
df['prices'].max()
```

## If Statement Using Loc()

```python
# All conditions using loc()

df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df['price_range_loc'].value_counts(dropna = False)
```

## For Loops

```python
# Finding busiest day of the week

df_ords_prod_merg['order_day_of_the_week'].value_counts(dropna = False)
```

```python
#Creating column to contain busiest days

result = []

for value in df_ords_prod_merg["order_day_of_the_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

## Problem 4

```python
# Looking at frequency of hours

df_ords_prod_merg['order_hour_of_day'].value_counts(dropna = False)
```

Looking at this we can break the hours into 3 groups, Most orders = [10, 11, 14, 15, 13, 12, 16, 9]. Least orders = [23, 6, 0, 1, 5, 2, 4, 3]. Lastly, Average orders = [17, 8, 18, 19, 20, 7, 21, 22]

```python
# Creating a for-loop if statement

hours = []

for value in df_ords_prod_merg["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    hours.append("Most orders")
  elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
    hours.append("Fewest orders")
  else:
    hours.append("Average orders")
```

```python
# Assigning new column to hours column

df_ords_prod_merg['busiest_period_of_day'] = hours
```
