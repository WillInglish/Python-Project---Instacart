## Problem 4

```python
# The three step process is:
# Split the data into groups based on the “user_id” column.
# Apply the transform() function on the “order_number” column to generate the maximum orders for each user.
# Create a new column, “max_order,” into which you’ll place the results of your aggregation.

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Creating loyalty flags based off of max_order column

ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# Verifying the flags are correctly assigned
ords_prods_merge[['user_id', 'loyalty_flag', 'order_number', 'max_order']].head(60)
```

## Problem 6

The criteria is:

    [1. If mean(prices) < 10, then flag = "Low spender"], 
    [2. If mean(prices) => 10, then flag = " High spender"]

```python
# Getting a column to find the averages of each user
ords_prods_merge['average_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)

ords_prods_merge.head()
```

```python
#Creating spending flags on criteria given using the average_price column
ords_prods_merge.loc[ords_prods_merge['average_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['average_price'] >=10, 'spending_flag'] = 'High spender'
```

```python
# Verifying by looking at the frequency of each flag.
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

## Problem 7

The criteria is:

    [1. If median(days since prior order) > 20, then flag = "Non-frequent customer"], 
    [2. If median(days since prior order) >10 AND <= 20, then flag = "Regular customer"],
    [3. If median(days since prior order) < 10, then flag = "Frequent customer"],

```python
# Creating a column of median days since prior order grouping by user_id and transforming using median

ords_prods_merge['median_prior_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)

ords_prods_merge.head()
```

```python
# Creating order_frequency_flag based on criteria

ords_prods_merge.loc[ords_prods_merge['median_prior_order'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_prior_order'] > 10) & (ords_prods_merge['median_prior_order']<= 20), 'order_frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_prior_order'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
# Checking the frequency of the new flags

ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```

```python
# Checking the null values

null_order = ords_prods_merge[ords_prods_merge['order_frequency_flag'].isnull()]

null_order
```
